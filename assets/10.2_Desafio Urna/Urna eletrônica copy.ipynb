### Objetivo:

### Descrição:

#### Na "Urna Eletrônica" que vamos desenvolver, cada grupo de alunos será responsável por construir um sistema simulado de votação. Os candidatos imaginários, e todos terão a oportunidade de exercer seu direito de voto em um ambiente virtual.

### Passos da Atividade:

#### Passo 2: Criando os Candidatos e Partidos

#### Passo 3: Simulando a Votação

#### Passo 4: Contagem e Anúncio dos Resultados

#### Ao encerrar a votação, exiba um relatório com os resultados da votação.

```python
candidatos = []

def cadastrar_candidatos():
    print('\n --------- CADASTRAR CANDIDATO --------- \n')
    candidato = input('Digite o nome do candidato: ').upper()
    if not candidato in candidatos:
        partido = input('Digite o partido do candidato: ').upper()
        numero = int(input('Digite o número correspondente ao candidato: '))
        candidato = {'candidato': candidato, 'partido': partido, 'numero': numero, 'votos': 0}
        candidatos.append(candidato)
        print('\n Candidato: {} \n Partido: {} \n Número: {} \n\n Candidato cadastrado com sucesso!!! \n'.format(candidato.get('candidato'), candidato.get('partido'), candidato.get('numero')))
    else:
        print('Candidato já cadastrado.')
    return mostrar_menu()

def votar_candidato():
    print('\n--------- ESCOLHA SEU CANDIDATO ---------\n')
    while True:
        numero = int(input('Digite o número do seu candidato: '))
        try:
            for candidato in candidatos:
                for chave, valor in candidato.items():
                    if candidato.get('numero') == numero:
                        candidato['votos'] += 1
                        print('\n Você votou em: \n')
                        print(' Candidato: {} \n Partido: {} \n Número: {}'.format(candidato.get('candidato'), candidato.get('partido'), candidato.get('numero')))
                        print('\n Obrigado pelo seu voto!!! \n')
                        break
        except:
            print('\n Você votou em nulo. \n')
        return mostrar_menu()
                 
def imprimir_votos():
    
    totalVotos = 0
    votos = 0
    vencedor = []

    print('\n--------- RELATÓRIO DE VOTOS ---------\n')
    for candidato in candidatos:
        totalVotos += candidato.get('votos')

    for candidato in candidatos:
        candidato['percentual'] = 0
        for chave, valor in candidato.items():
            percentual = round((candidato.get('votos') / totalVotos) * 100)
            candidato['percentual'] = percentual 
            print(' Candidato: {} \n Partido: {} \n Número de votos: {} \n Percentual: {} %\n'.format(candidato.get('candidato'), candidato.get('partido'), candidato.get('votos'), candidato.get('percentual')))
            if candidato.get('votos') > votos:
                votos = candidato.get('votos')
                vencedor = candidato
            break
        
    print('*** VENCEDOR *** \n')
    print(' Candidato: {} \n Partido: {} \n Numero: {} \n Total de votos: {} \n Percentual: {} %'.format(vencedor.get('candidato'), vencedor.get('partido'), vencedor.get('numero'), vencedor.get('votos'), vencedor.get('percentual')))

    return mostrar_menu()

def direcionar_opcoes(opcao):
    
    if opcao != '':
        if opcao == 1: cadastrar_candidatos()
        elif opcao == 2: votar_candidato()
        elif opcao == 3: imprimir_votos()
        elif opcao == 4: 
            print('\n --------- Programa finalizado ---------\n')
        else:
            print('\nOpção inválida.')
            mostrar_menu()
    else:
        print('Favor, selecionar uma opção.')
        
def mostrar_menu():
    print('\n --------- MENU DE OPÇÕES --------- \n')
    print('Opção 1 - Cadastrar candidato')
    print('Opção 2 - Votar')
    print('Opção 3 - Imprimir resultado')
    print('Opção 4 - Sair do programa')
    opcao = int(input('Digite sua opção: '))
    direcionar_opcoes(opcao)

mostrar_menu()
```
