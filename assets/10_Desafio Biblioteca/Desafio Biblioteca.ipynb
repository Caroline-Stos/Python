## Tarefas:

## 1 - Registro de Empréstimos:

### Registre os empréstimos de livros para os  alunos fictícios da escola.
## 2 -  Avaliação de Livros:

### Após os empréstimos, simule que os alunos leram os livros e atribua notas de avaliação aos livros emprestados.

## 3 - Avaliação do Atendimento da Biblioteca:

###  Registre avaliações do atendimento da biblioteca por parte dos alunos.

## 4 - Imprimir Relatório:

### Após realizar as ações anteriores, imprima um relatório que apresente as informações sobre os empréstimos, avaliações dos livros e a média das avaliações do atendimento da biblioteca.


```python
alunos = {}

alunos[('ANA')] = {'livro': 0, 'nota_livro': 0, 'nota_atend': 0}
alunos[('CARLOS')] = {'livro': 0, 'nota_livro': 0, 'nota_atend': 0}
alunos[('MARCIA')] = {'livro': 0, 'nota_livro': 0, 'nota_atend': 0}

livros = []

# funcao para emprestar o livro
def emprestar_livro(alunos, aluno, livro):
    if aluno in alunos:
        if not livro in livros:
            alunos[aluno]['livro'] = livro
            livros.append(livro)
        else:
            print('Este livro já está em empréstimo.')
    else:
        print('Aluno não encontrado.')
    return alunos, livros

# funcao para avaliar o livro
def avaliar_livro(alunos, aluno, livro, nota):
    if aluno in alunos:
        if livro in livros:
            alunos[aluno]['nota_livro'] = nota
        else:
            print('Este livro não está em empréstimo.')
    else:
        print('Aluno não encontrado.')
    return alunos

# funcao para avaliar o atendimento
def avaliar_atendimento(alunos, aluno, nota):

    if aluno in alunos:
        alunos[aluno]['nota_atend'] = nota
    else:
        print('Aluno não encontrado.')
    return alunos

# funcao para calcular media e imprimir relatório
def relatorio(alunos):

    notas = []
    for aluno in alunos:
        nota = alunos[aluno].get('nota_atend')
        if nota != 0:
            notas.append(nota)
    
    if len(notas) != 0:
        media = sum(notas) / len(notas)
        media = round(media)
    else:
        media = 'Sem notas suficientes para calcular média.'

    print(" RELATÓRIO DE EMPRÉSTIMOS: ")
    
    for aluno in alunos:
        print(f"""
               
        Aluno: {aluno}
        Livro: {alunos[aluno]['livro']}
        Nota - livro: {alunos[aluno]['nota_livro']}
        Nota - Atendimento: {alunos[aluno]['nota_atend']}
        """)
        
    print(F'Média de atendimento: {media}')

# while de opções para selecionar
while True:
    options = int(input("""Selecione uma das opções abaixo:
                    1 - Emprestar Livro.
                    2 - Avaliar Livro.
                    3 - Avaliar Atendimento.
                    4 - Imprimir relatório.
                    """))
                    
    if options == 1:
        aluno = input('Digite o nome do aluno: ').upper()
        livro = input('Digite o nome do livro: ').upper()
        emprestar_livro(alunos, aluno, livro)

    if options == 2:
        aluno = input('Digite o nome do aluno: ').upper()
        livro = input('Digite o nome do livro: ').upper()
        nota = int(input('Digite a nota de avaliação do livro (0 a 10): '))
        if nota in range(0, 11):
            avaliar_livro(alunos, aluno, livro, nota)
        else:
            print('Nota inválida, informe uma nota de 0 a 10.')
    
    if options == 3:
        aluno = input('Digite o nome do aluno: ').upper()
        nota = int(input('Digite a nota de avaliação do atendimento (0 a 10): '))
        if nota in range(0, 10):
            avaliar_atendimento(alunos, aluno, nota)
        else:
            print('Nota inválida, informe uma nota de 0 a 10.')
        
    if options == 4:
        relatorio(alunos)
```
