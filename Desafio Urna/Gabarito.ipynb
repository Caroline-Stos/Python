```python
candidatos = []
voto = ""

def verifica_disponibilidade(numero):
    for candidato in candidatos:
        for chave,valor in candidato.items():
            if (chave == "numero") and (valor == numero):                  
                print("NUMERO JÁ EXISTE NA BASE DE CANDIDATOS. CADASTRO CANCELADO")
                return False
            else :
                return True       

def cadastra_candidato():
   print("****** CADASTRO DE CANDIDATO ********")
   try:
        numero = input("Digite o numero do candidato: ")       
        if candidatos == []: 
            pode_cadastrar = True
        else:
            pode_cadastrar = verifica_disponibilidade(numero)    

        if pode_cadastrar:
                print("NUMERO DISPONÍVEL, PROSSEGUINDO CADASTRO..")
                nome = input("Nome do Candidato: ")
                partido = input("Digite o partido: ")
                novo_candidato = {"numero": numero, "nome": nome, "partido": partido, "votos": 0}
                candidatos.append(novo_candidato)
                print("\n *** CANDIDATO CADASTRADO ***\n")                
   except:
        print("\n *** erro ao cadastrar - tente novamente ***\n")

def analisa_opcao(opcao):
    if opcao == "3":
        print("Programa Finalizado")
    elif (opcao=="1") or (opcao=="2"):
        executa_opcao(opcao)
    else:
        print('\n*** opcao invalida ***\n')

def executa_opcao(opcao):
    if(opcao == "1"): cadastra_candidato()
    if(opcao == "2"): inicia_votacao()  

def imprime_relatorio(total_votos, maior_votacao, vencedor, nulos):
    print("\n********** ELEIÇÕES 2023 ***************")
    print("********* RESULTADO OFICIAL **************")
    print("********* VOTOS ************")
    if(total_votos != 0) and (total_votos != nulos):
        for item in candidatos:
            for chave,valor in item.items():
                print(str(chave) + "  :  " + str(valor))            
            print("-------------------------------\n")
            
            if(item.get("votos") > maior_votacao):
                    maior_votacao = item.get("votos")
                    vencedor = item

        percentual_vencedor = ( vencedor.get("votos") / (total_votos - nulos) ) * 100  
        print(f"Total de Votos: {total_votos}")
        print(f"Votos Nulos: {nulos}")
        print("Parabéns ao vencedor {} com {:.2f}% dos votos válidos!".format(vencedor.get("nome"),percentual_vencedor))    
        print("*************************************\n")
        print("********** FIM DO RELATÓRIO  ********\n")
        print("*************************************\n")
        return
    else:
        print("não há votos válidos, reiniciando programa...\n********************************".upper())    

def inicia_votacao():       
    voto = 0
    total_votos = 0
    nulos = 0
    maior_votacao = 0
    vencedor = {}  
    while voto != "SENHAXYZ":
        print("\n********* VOTAR **********")
        voto = input('>>>> DIGITE O NUMERO DO CANDIDATO: ')        
        if voto == "SENHAXYZ":
            imprime_relatorio(total_votos, maior_votacao, vencedor, nulos)
            return
        else:
            candidato_escolhido = next((c for c in candidatos if c["numero"] == voto), None)
            if candidato_escolhido != None:
                print(f"Candidato: {candidato_escolhido.get('nome')} Partido: {candidato_escolhido.get('partido')}")                
                confirma = input("Pressione C maiúsculo para confirmar, outra tecla para reiniciar: ")
                if confirma != "C": 
                    print("VOTO NÃO REGISTRADO - REINICIANDO...")
                    continue
                candidato_escolhido["votos"] += 1
                total_votos += 1
                print("VOTO REGISTRADO")
            else:
                print("CANDIDATO INEXISTENTE")
                confirma = input("Pressione C maiúsculo para confirmar VOTO NULO, outra tecla para reiniciar: ")
                if confirma != "C": 
                    print("VOTO NÃO REGISTRADO - REINICIANDO...")
                    continue
                total_votos += 1               
                nulos += 1

def mostra_menu():
    print('''***** ELEIÇÕES 2023 - Digite a opção e tecle ENTER: *****

        1. Cadastro de Candidato
        2. Iniciar Votação
        3. Finalizar Programa''')

def programa_principal():
    opcao = 0 
    while(opcao != "3"):                     
            mostra_menu()
            opcao = input('''
            >>>>>>>>>>> Digite a opção desejada:  ''')
            analisa_opcao(opcao)
    
programa_principal()
```
