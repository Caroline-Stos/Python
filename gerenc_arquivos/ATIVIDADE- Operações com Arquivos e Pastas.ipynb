#### Exercício 4:
Crie um programa que verifique se um arquivo existe em um diretório usando os.path.exists()

```python
from pathlib import Path

def verificando_arq(caminho):
    arquivo = input('Digite o nome do arquivo completo:')
    if (caminho/Path(arquivo)).exists(): # verifica se um arquivo existe
         print('Este arquivo já existe.')
    else:
        print('Este arquivo não existe.')
    
caminho = Path.cwd() # tras o caminho da pasta atual
verificando_arq(caminho) # chamando funcao
```

#### Exercício 5:
Crie um programa que leia o conteúdo de um arquivo e o imprima na tela usando open()

```python
from pathlib import Path

def ler_arquivo(caminho):
    try:
        arquivo = input('Digite o nome do arquivo completo com extensão:')
        if (caminho/Path(caminho/arquivo)).exists():
            arquivo = open(arquivo,'r') # abri o arquivo para leitura
            ler = arquivo.readlines() # le as linhas do arquivo de texto
            print(ler)
    except:
        print('Este arquivo não existe.')

caminho = Path.cwd() # tras o caminho da pasta atual
ler_arquivo(caminho) # chamando funcao
```

#### Exercício 6:
Crie um programa que escreva uma lista de strings em um arquivo de texto usando open()

```python
# gravar nome e email em um arquivo txt

pacientes = []

def cadastrando_pacientes(paciente):
    with open('paciente_ex6','a') as arquivo: # abre o arquivo e 
        for paciente in pacientes:
            arquivo.write(str(paciente))

while True:
    nome = input('Digite o nome do paciente:')
    if nome == '':
        cadastrando_pacientes(paciente)
        break
    email = input('Digite o e-mail do paciente:')
    paciente = [nome, email]
    pacientes.append(paciente)
```

#### Exercício 7:
Crie um programa que copie um arquivo de um diretório para outro usando shutil.copy()

```python

import shutil as stl
from pathlib import Path

def copiar_arquivo():
    copiar = Path(input('Origem do arquivo a ser copiado: '))
    colar = Path(input('Destino do arquivo a ser colado: '))
    stl.copy2(copiar, colar) # copia e cola um arquivo

copiar_arquivo()
```

#### Exercício 8:
Crie um programa que renomeie um arquivo usando os.rename()

```python
import os

def renomear_arquivo():
    try:
        nome_atual = input('Digite o nome do arquivo que deseja renomear: ')
        novo_nome = input('Novo nome do arquivo: ')
        os.rename(nome_atual, novo_nome) # renomeia um arquivo
        print(f'Arquivo renomeado com sucesso para: {novo_nome}!')
    except:
        print('Error: Não foi possível renomear o arquivo. Verifique e tente novamente.')

renomear_arquivo()
```
